## Define Time Calculation Class Functions

```python
class Timer:
    def __init__(self):
        self._running = False
        self._start_time = None
        self._end_time = None
        self._total_time = None
    def start(self):
        if self._running:
            return False
        self._start_time = datetime.datetime.now()
        self._running = True
        return self._start_time
    def end(self):
        if self._running:
            self._end_time = datetime.datetime.now()
            self._running = False
            self._total_time = self._end_time - self._start_time
            return self._total_time
    def get_current(self):
        if self._running:
            current_time = datetime.datetime.now()
            process_time = current_time - self._start_time
            return process_time
        else:
            return 0
    def running(self):
        return self._running
    def get_total_time(self):
        if self._total_time:
            return self._total_time
        return False
```

# Define  Plot History Function

```python
def show_history_plots(history):
    # summarize history for accuracy
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
```

## Defining Model Creation Function

```python
def create_cnn_model(input_shape=(64, 64, 3), batch_size=256):
    #batch_input_shape = (batch_size, input_shape[1], input_shape[2], input_shape[3])
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Defining the model
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Model: [conv-relu-conv-relu-pool]xN -> [affine]xM -> [softmax or SVM]
    # --####-----------------------------------------------------------------------
    # --#### Defines a sequential model
    new_model = tf.keras.Sequential()
    #new_model.add(tf.keras.layers.InputLayer(input_shape=input_shape, batch_size=batch_size))

    # --####--------------------------------------------
    # --#### Conv -> Conv -> ReLU -> MaxPool -> Dropout
    # --####--------------------------------------------
    # --#### Adds a Convolution layer without an activation function to the model
    #new_model.add(tf.keras.layers.Conv2D(128, kernel_size=(5, 5), strides=(1, 1), padding="same", input_shape=input_shape))
    # --#### Adds the activation function to the model
    #new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    #new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    #new_model.add(tf.keras.layers.Dropout(.5))
    # --#### Adds a Convolution layer without an activation function to the model
    #new_model.add(tf.keras.layers.Conv2D(64, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds the activation function to the model
    #new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    #new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    #new_model.add(tf.keras.layers.Dropout(.5))



    # --####--------------------------------------------
    # --#### Conv -> Conv -> ReLU -> MaxPool -> Dropout
    # --####--------------------------------------------
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(16, kernel_size=(5, 5), strides=(1, 1), padding="same", input_shape=input_shape))
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(16, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds the activation function to the model
    new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    new_model.add(tf.keras.layers.Dropout(.5))


    # --####--------------------------------------------
    # --#### Conv -> Conv -> ReLU -> MaxPool -> Dropout
    # --####--------------------------------------------
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(32, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(32, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds the activation function to the model
    new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    new_model.add(tf.keras.layers.Dropout(.5))
    # --####--------------------------------------------
    # --#### Conv -> Conv -> ReLU -> MaxPool -> Dropout
    # --####--------------------------------------------
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(64, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(64, kernel_size=(5, 5), strides=(1, 1), padding="same"))
    # --#### Adds the activation function to the model
    new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    new_model.add(tf.keras.layers.Dropout(.5))
    # --####--------------------------------------------
    # --#### Conv -> Conv -> ReLU -> MaxPool -> Dropout
    # --####--------------------------------------------
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(128, kernel_size=(3, 3), strides=(1, 1), padding="same"))
    # --#### Adds a Convolution layer without an activation function to the model
    new_model.add(tf.keras.layers.Conv2D(128, kernel_size=(3, 3), strides=(1, 1), padding="same"))
    # --#### Adds the activation function to the model
    new_model.add(tf.keras.layers.Activation(tf.nn.relu))
    # --#### Adds a pooling layer to the model
    new_model.add(tf.keras.layers.MaxPool2D(pool_size=(2, 2)))
    # --#### Adds Dropout to the model
    new_model.add(tf.keras.layers.Dropout(.5))
    # --####-----------------------------------------------------------
    # --#### Flatten -> ReLU -> Dropout -> ReLU -> Dropout -> sigmoid
    # --####-----------------------------------------------------------
    # --#### Adds a flatten layer to the model (Flattens the output)
    new_model.add(tf.keras.layers.Flatten())
    # --#### Adds a dense hidden layer of the specified number of neurons
    new_model.add(tf.keras.layers.Dense(units=1024, activation="relu"))
    # --#### Adds a dropout with 50% chance of dropping a value
    new_model.add(tf.keras.layers.Dropout(rate=0.3))
    # --#### Adds a dense hidden layer of the specified number of neurons
    new_model.add(tf.keras.layers.Dense(units=10, activation="relu"))
    # --#### Adds a dropout with 50% chance of dropping a value
    new_model.add(tf.keras.layers.Dropout(rate=0.1))
    # --#### Adds the output layer of the specified number of neurons
    new_model.add(tf.keras.layers.Dense(units=2, activation="sigmoid"))
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Returns the created model
    return new_model
```

## Defining Building Model Function With Passed Hyperparameters To Be Tuned

```python
def build_test_model(model, loss_optimize = 'categorical', lrn_rate: float = 0.0001, batch_size=25, image_size_1=64, image_size_2=64, channels=3) -> None:
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Building, Defining the Optimizer, Compiling & Showing Model Summery
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Builds the Model
    model.build(input_shape=[batch_size, image_size_1, image_size_2, channels])
    # --#### Define Optimizer
    optimizer = tf.optimizers.Adam(learning_rate=lrn_rate)
    # --#### Compiles the model
    if loss_optimize == 'categorical':
        use_loss_optimizer = tf.keras.losses.SparseCategoricalCrossentropy
        model.compile(loss=use_loss_optimizer(), metrics=['accuracy'], optimizer=optimizer)
    elif loss_optimize == 'binary':
        ROC = tf.keras.metrics.AUC()
        use_loss_optimizer = tf.keras.losses.BinaryCrossentropy
        model.compile(loss=use_loss_optimizer(), metrics=['accuracy', ROC], optimizer=optimizer)
    # --#### Shows information about the Model
    ########model.summary()
    return model
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #

def build_test_model_5(model, loss_optimize = 'categorical', lrn_rate: float = 0.0001, batch_size=25, image_size_1=96, image_size_2=96, channels=3) -> None:
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Building, Defining the Optimizer, Compiling & Showing Model Summery
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
    # --#### Builds the Model
    model.build(input_shape=[batch_size, image_size_1, image_size_2, channels])
    # --#### Define Optimizer
    optimizer = tf.optimizers.Adam(learning_rate=lrn_rate)
    # --#### Compiles the model
    if loss_optimize == 'categorical':
        use_loss_optimizer = tf.keras.losses.SparseCategoricalCrossentropy
        model.compile(loss=use_loss_optimizer(), metrics=['accuracy'], optimizer=optimizer)
    elif loss_optimize == 'binary':
        ROC = tf.keras.metrics.AUC()
        use_loss_optimizer = tf.keras.losses.BinaryCrossentropy
        model.compile(loss=use_loss_optimizer(), metrics=['accuracy', ROC], optimizer=optimizer)
    # --#### Shows information about the Model
    ########model.summary()
    return model
    # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #
```

```python
# --#### Creates an overall tuning timer
overall_timer = Timer()
# --#### Starts the overall tuning Timer's Timeing
overall_start_time = overall_timer.start()
# --#### Creates a list to store all the models and their results
results = []
# --#### Defines the number of Epochs to use
epochs = 25
# --#### Defines the batch size
batch_size = 50
# --#### Learning Rate
lrn = None

# --#### Convert Image data to tensor
#X = tf.convert_to_tensor(X_train, dtype=np.float32)
# --#### Convert Correct Image label to tensor
#Y = tf.convert_to_tensor(Y_train, dtype=np.float32)
# --#### Convert Image data to tensor
#Xtest = tf.convert_to_tensor(X_test, dtype=np.float32)
# --#### Convert Correct Image label to tensor
#Ytest = tf.convert_to_tensor(Y_test, dtype=np.float32)

count = 0
while epochs <= 125:  # --#### Goes over Epochs 50 - 150
    epochs += 25
    # --#### Defines the batch size
    batch_size = 50
    while batch_size <= 250:  # --#### Goes over Epochs 100 - 300
        batch_size += 50
        count += 1

        results.append([])
        # --#### Runs parameter test 3x ,Due to random dropouts parameters don't give same results.
        for cnt in range(2):
            dl()
            cp(
                f"Testing Model {color_start}{count}.{cnt+1}{color_stop}: Epochs={color_start}{epochs}{color_stop}, Batch Size={color_start}{batch_size}{color_stop}, lrn={color_start}{lrn}{color_stop}",
                directory_colors,
                use_time_stamp=False,
                color=[value_color, value_color, value_color, value_color],
                lev=3
            )
            dl(25)
            # --#### Creates a hyperparameter tuning model
            tune_model = create_cnn_model()
            # --#### Builds the hyperparameter tuning model
            build_test_model(tune_model)

            # --#### Creates a Time object to calculate total time spent training the model
            timer = Timer()
            # --#### Starts the Timer's Timeing
            start_time = timer.start()
            # --#### Trains the model
            history = tune_model.fit(train_generator, epochs=epochs, batch_size=batch_size, verbose=1, validation_data = valid_generator)
            # --#### Stops the Timer's Timeing and obtains the total timer run time (Time spent training the model)
            total_time = timer.end()
            cp(f"Total Training Time: {color_start}{total_time}{color_stop}", color=value_color)
            dl(100)
            cp(f"Evaluating Model Generalization", use_time_stamp=False)
            # --#### Evaluates how well the model generalizes
            eval_vals = tune_model.evaluate(valid_generator)
            dl(100)
            # --#### Stores and appends the results of the test
            store_info = [count, eval_vals[-1], eval_vals[0], [epochs, batch_size, lrn, eval_vals, tune_model, history]]
            results[-1].append(store_info)
            cp(f"Total Processing Time: {color_start}{overall_timer.get_current()}{color_stop}", color=value_color, use_time_stamp=False)
            # --#### --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ####-- #

best_accuracy = 0.0
best_loss = 1000
best_result_output = None
for val in results:
    for val2 in val:
        cp(val2, value_color, use_time_stamp=False, lev=2)
        if val2[1] >= best_accuracy and val2[2] <= best_loss:
            best_accuracy = val2[1]
            best_loss = val2[2]
            best_result_output = val2

cp(f"Best accuracy: {best_accuracy}, Best loss: {best_loss}\n\t{best_result_output}", use_time_stamp=False)
# --#### Stops the Timer's Timeing and obtains the total timer run time (Time spent training the model)
overall_total_time = overall_timer.end()
cp(f"Total Tuning Time: {color_start}{overall_total_time}{color_stop}", color=value_color)
```
